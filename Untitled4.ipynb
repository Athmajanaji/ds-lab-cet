```python
import matplotlib.pyplot as mp
import numpy as np

xpoints = np.array([1,2,6,18])
ypoints = np.array([3,10,12,20])

mp.plot(xpoints,ypoints,'o:r',mfc="g",mec="g")
mp.show()
```

```python
import matplotlib.pyplot as mp
import numpy as np

with open("file21.txt") as f:
    r = f.read()

#print(r)

r=r.split('\n')

#print(r)

x = [row.split(' ')[0] for row in r]
y = [row.split(' ')[1] for row in r]

mp.plot(y,x,'o-r',mfc="g",mec="g")
mp.xlabel('x-axis')
mp.ylabel('y-axis')

mp.show()
```
